```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re

def translate_sentence(sentence):
    # URL encode the sentence
    url_encoded_sentence = sentence.replace(' ', '%20')
    
    # Construct the URL
    url = "https://itranslate.com/translate/english-to-yoruba/"+url_encoded_sentence
    
    # Fetch HTML content
    driver.get(url)
    
    # Wait until the translation is loaded
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "targetText")))
    
    # Fetch the HTML content after the translation is loaded
    html_content = driver.page_source
    
    # Extract relevant information from HTML content
    title_match = re.search(r'<title>(.*?) in Yoruba - iTranslate</title>', html_content)
    original_sentence = title_match.group(1).strip() if title_match else None
    
    translation_match = re.search(r'<div class="targetText" role="presentation">(.*?)</div>', html_content)
    translated_sentence = translation_match.group(1).strip() if translation_match else None
    
    return original_sentence, translated_sentence

# Initialize WebDriver
driver = webdriver.Chrome()

# List of sentences to translate
sentences_to_translate = ["how are you" , "sunset at beach"]

# Translate sentences
translated_sentences = []
for sentence in sentences_to_translate:
    original, translated = translate_sentence(sentence)
    translated_sentences.append((original, translated))

# Close WebDriver

# Process translated sentences
for original, translated in translated_sentences:
    print("Original Sentence:", original)
    print("Translated Sentence in Yoruba:", translated)
    print()

driver.quit()

# Handle any errors that occurred during translation process
```

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re

def translate_sentence(sentence):
    # URL encode the sentence
    url_encoded_sentence = sentence.replace(' ', '%20')
    
    # Construct the URL
    url = "https://itranslate.com/translate/english-to-yoruba/"+url_encoded_sentence
    
    # Fetch HTML content
    driver.get(url)
    
    # Wait for some time for the page to load
    time.sleep(3)  # Adjust the time as needed
    
    # Fetch the HTML content
    html_content = driver.page_source
    
    # Extract relevant information from HTML content
    title_match = re.search(r'<title>(.*?) in Yoruba - iTranslate</title>', html_content)
    original_sentence = title_match.group(1).strip() if title_match else None
    
    translation_match = re.search(r'<div class="targetText" role="presentation">(.*?)</div>', html_content)
    translated_sentence = translation_match.group(1).strip() if translation_match else None
    
    return original_sentence, translated_sentence

# Read the CSV file
input_csv_file = 'Image_Generation_Prompts.csv'
output_csv_file = 'translated_sentences.csv'
df = pd.read_csv(input_csv_file)

total_sentences = len(df['eng'])
processed_sentences = 0

# Initialize WebDriver
driver = webdriver.Chrome()

# Translate sentences
translated_sentences = []
for i, sentence in enumerate(df['eng'], 1):
    original, translated = translate_sentence(sentence)
    translated_sentences.append([original, translated])
    
    # Print progress
    processed_sentences += 1
    print(f"Processed {processed_sentences}/{total_sentences} sentences ({i}/{total_sentences}):")
    print("English sentence:", original)
    print("Translated sentence:", translated)
    print()
    
    # Close and reopen driver after every 50 sentences
#     if i % 50 == 0:
#         driver.quit()
#         driver = webdriver.Chrome()

# Save translated sentences to a new CSV file
translated_df = pd.DataFrame(translated_sentences, columns=['org_sentence', 'trans_sentence'])
translated_df.to_csv(output_csv_file, index=False)

# Close the last WebDriver instance
driver.quit()
```
