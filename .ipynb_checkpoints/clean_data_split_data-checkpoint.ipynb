```python
import pandas as pd
```

```python
raw_data = pd.read_csv(r'translated_sentences_eng_yor_1300.csv')
```

```python
raw_data = raw_data.drop_duplicates(subset='eng')
raw_data = raw_data.drop_duplicates(subset='yor')

display(raw_data)
```

```python
import pandas as pd

# Assuming raw_data is your DataFrame
# Replace 'column1' and 'column2' with the actual names of your columns
# Assuming the text is stored as strings in the DataFrame

# Count occurrences of <div> in column1
count_column1 = raw_data['eng'].str.contains('<div').sum()

# Count occurrences of <div> in column2
count_column2 = raw_data['yor'].str.contains('<div').sum()

# Total count across both columns
print("colum_1 :", count_column1)
print("colum_2 :", count_column2)

total_count = count_column1 + count_column2

print("Total number of rows containing '<div>' text:", total_count)
```

```python

# Filter out rows containing '<div>' text in either column
cleaned_data = raw_data[~(raw_data['eng'].str.contains('<div') | raw_data['yor'].str.contains('<div'))]


cleaned_data['eng'] = cleaned_data['eng'].apply(lambda x: x[:-1] if x.endswith('.') else x)

cleaned_data['yor'] = cleaned_data['yor'].apply(lambda x: x[:-1] if x.endswith('.') else x)



# Display the cleaned DataFrame
display(cleaned_data)
```

```python
# Iterate over each row in the DataFrame
for index, row in cleaned_data.iterrows():
    # Print the content of each row
    print(row['eng'])
    print(row['yor'])
    print()
```
